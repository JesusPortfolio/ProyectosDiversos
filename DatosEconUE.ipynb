```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Accediendo a los datos

```python
dt1=pd.read_excel('Indicadores_banco_mundial.xlsx')
dt1.head()
```

```python
dt1.set_index('Country Name',inplace=True)
```

```python
dt2=dt1.drop(columns=['Series Code'])
dt2.head(10)
```

# Limpieza de datos

```python
columnas = ['Series Name','Country Code'] + [str(k) for k in range(2000, 2016)]
dt3 = dt2.copy()
dt3.columns = columnas
dt3
```

```python
paises=['Germany','Spain','Portugal','Belgium']
dt4=dt3.loc[paises]
dt4.head()
```

```python
dt5=dt4.loc[dt4['Series Name']=='Inflation, consumer prices (annual %)']#Aqui se pueden seleccionar
#lineas con dos datos de cada linea
dt5.head()
```

```python
dt6=dt5.drop(columns=['Series Name','Country Code'])
dt6.head()
```

# Visualizaciones

```python
plt.figure(figsize=(10,6))
for pais in paises:
    plt.plot(dt6.loc[pais],label=pais)
plt.grid()
plt.legend()
plt.title('Inflation, consumer prices (annual %)')
plt.xlabel('Years')
plt.ylabel('Inflation (annual %)')
plt.show()

```

```python
dt6.mean(axis =1)
```

```python
plt.figure(figsize=(10,6))
plt.title('Inflation mean for the period')
plt.pie(dt6.mean(axis =1),labels=dt6.index,autopct='%1.1f%%')
plt.show()
```
