```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Acediendo a los datos

```python
dt1=pd.read_excel('Indicadores_banco_mundial.xlsx')
dt1
```

```python
dt1.set_index('Country Name',inplace=True)
```

```python
dt1.head()
```

```python
dt2=dt1.drop(columns=['Series Code'])
dt2.head()
```

```python
colunas=['Series Name','Country Code']+[str(k) for k in range(2000,2016)]
```

```python
dt3=dt2.set_axis(colunas,axis=1,inplace=False)#cambie todos los nombres de 
#las colunas!!!
dt3
```

```python
paises=['Germany','Spain','Portugal','Belgium']
dt4=dt3.loc[paises]
dt4.head(50)
```

```python
dt5=dt4.loc[dt4['Series Name']=='Inflation, consumer prices (annual %)']#Aqui se pueden seleccionar
#lineas con dos datos de cada linea
dt5.head()
```

```python
dt6=dt5.drop(columns=['Series Name','Country Code'])
dt6.head()
```

```python
plt.figure(figsize=(12,6))
for pais in paises:
    plt.plot(dt6.loc[pais],label=pais)
plt.grid()
plt.legend()
plt.title('Inflation, consumer prices (annual %)')
plt.xlabel('Years')
plt.ylabel('Inflation (annual %)')
plt.show()

```

```python
dt6.mean(axis =1)
```

```python
plt.figure(figsize=(12,6))
plt.title('Inflation mean for the period')
plt.pie(dt6.mean(axis =1),labels=dt6.index,autopct='%1.1f%%')
plt.show()
```

```python
dt7=dt4.loc[dt4['Series Name']=='Exports of goods and services (% of GDP)'],['Imports of goods and services (% of GDP)']]
dt7.head()
```

```python
dt8=dt7.drop(columns=['Series Name','Country Code'])
```

```python
plt.figure(figsize=(12,6))
for pais in paises:
    plt.plot(dt8.loc[pais],label=pais)
    plt.legend()
    plt.title('Exports of goods and services (% of GDP)')
    plt.xlabel('Years')
    plt.ylabel('% of GDP')
plt.grid()
plt.show()
```
